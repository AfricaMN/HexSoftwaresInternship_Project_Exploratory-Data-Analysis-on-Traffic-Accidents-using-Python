```python
#Data Cleaning
```

```python
#Make a copy of the original data so that we can use it antime we want
data_copy = data
```

```python
#Check null values from the data
print("\nDisplay null values from the data:")
data_copy.isnull().sum()
```

```python
#Dealing with duplicates data
data_copy.duplicated().sum() 
data_copy = data_copy.drop_duplicates() # drop duplicated data
data_copy.duplicated().any()
```

```python
#Handling missing values
print("\nHandling missing values")
data_copy = data_copy.replace('UNKNOWN',np.nan) #replace all unknown values with null so i can dirctly handle it
missing = data_copy.isnull().sum()/data_copy.shape[0]*100
missing
```

```python
#Remove unwanted data
```

```python
traffic_accidents = data_copy.drop(['road_defect','intersection_related_i'], axis=1)
```

```python
traffic_accidents
```

#Data Visualisation

```python
numeric_columns = traffic_accidents.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = traffic_accidents[numeric_columns].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
#Traffic accidents caused by weather conditions
print("\nTraffic accidents caused by weather conditions: ")
data_copy['weather_condition'].value_counts().plot(kind='bar', title='Accidents caused by Weather Conditions')
```

```python
sns.countplot(x='weather_condition', hue='most_severe_injury', data=data_copy)
plt.title('Accidents by Weather Condition and Injury Severity')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='roadway_surface_cond', hue='crash_type', data=data_copy)
plt.title('Accidents by Road Surface Condition')
plt.xticks(rotation=45)
plt.show()
```

```python
data_copy['most_severe_injury'].value_counts().plot(kind='bar', title='Most Severe Injury')
```

```python
data_copy['crash_date'] = pd.to_datetime(data_copy['crash_date'])  # Ensure correct datetime format
data_copy['year_month'] = data_copy['crash_date'].dt.to_period('M')  # Extract year-month
data_copy.groupby('year_month').size().plot(kind='line', figsize=(10, 5), marker='o')
plt.title("Monthly Trend of Crashes")
plt.xlabel("Date (Year-Month)")
plt.ylabel("Number of Crashes")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
